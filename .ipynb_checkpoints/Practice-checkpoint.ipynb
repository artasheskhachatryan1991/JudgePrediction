```python
def add_ones(X):
    return np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
```

```python
def normalize_features(y):
    return(y-np.min(y, axis=0))/(np.max(y, axis=0) - np.min(y, axis=0))
```

```python
def mean_square_loss(X, Y, b):
    """
    :param X: A numpy matrix, where each row is a data element (X)
    :param Y: A list of responses for each of the rows (y)
    :param b: An array of parameters
    :return: mean square loss is square loss per data point
    """
    return np.sqrt((Y - X.dot(b)).T.dot(Y - X.dot(b)).sum()/X.shape[0])
```

```python
def fit_linear_regression(X, Y, Lambda):
    """
    :param X: A numpy matrix, where each row is a data element (X)
    :param Y: A list of responses for each of the rows (y)
    :return: An array of parameters
    """
    return pinv(X.T.dot(X) + Lambda * np.eye(X.shape[1])).dot(X.T).dot(Y)
```

```python
def cross_validation(X, Y, Lambda=0, n_splits=10):    
    loss = 0
    for i in range(n_splits):
        X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=1/n_splits)       
        b = fit_linear_regression(X_train, y_train, Lambda)
        loss += mean_square_loss(X_test, y_test, b)
    return loss/n_splits
```
